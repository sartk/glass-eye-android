```python
i = 0
a = []
for f in os.listdir(image_path+'Annotation'):
    a.append(f[10:f.index('.')])
print(a)
for f in os.listdir(image_path+'img'):
    n = f[10:f.index('.')]
    if n not in a:
        os.remove(image_path+'img/'+f)
        print(f)
        i-=1
    i+=1
    
print(i)

b = []
for f in os.listdir(image_path+'img'):
    b.append(f[10:f.index('.')])
i=0
for f in os.listdir(image_path+'Annotation'):
    n = f[10:f.index('.')]
    if n not in b:
        os.remove(image_path+'Annotation/'+f)
        print(f)
        i-=1
    i+=1
print(i)
```

```python
test_path = 'C:tensorflow1/models/research/object_detection/images/test/'
train_path = 'C:tensorflow1/models/research/object_detection/images/train/'
i = 0
tt_split = .85
size = len(os.listdir(image_path+'img'))
imax = size * tt_split
for f in os.listdir(image_path+'img'):
    if i<imax:
        os.rename(image_path+'img/'+f,train_path+f)
    else:
        os.rename(image_path+'img/'+f,test_path+f)
    i+=1
    
c = []
for f in os.listdir(train_path):
    c.append(f[10:f.index('.')])
for f in os.listdir(test_path):
    c.append(f[10:f.index('.')])
    
for f in os.listdir(image_path+'Annotation'):
    if f[10:f.index('.')] in c:
        os.rename(image_path+'Annotation/'+f,train_path+f)
    else:
        os.rename(image_path+'Annotation/'+f,test_path+f)
    
```
